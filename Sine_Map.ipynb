```python
import numpy as np
import sys
import matplotlib.pyplot as plt
import math
%matplotlib inline
plt.rcParams['figure.figsize'] = (8,6)
```

```python
# Sine Maps xt+1 = R/4 * (pi * xt)
math.pi
```

```python
R = float(input ('what is the value of R0?' ))
X = float(input ('what is the vale of x?'))
t = int(input ('How my interations(t) ?' ))
```

Creating a function to run the logic map formula

```python
def Logic_Map(X , R, t):
    
    out = np.array([])
    while t > 0:
        X = (R / 4) * math.sin(math.pi * X)
        out = np.append(out, [X])
        t = t - 1
         
    return out
```

```python
graph = (Logic_Map(X, R, t))
print (graph)
```

```python
plt.plot(graph)
plt.title('Sine Map')
plt.ylabel('X(t)')
plt.xlabel('t')
plt.show
```
